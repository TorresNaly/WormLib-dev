### 1. Interface Overview

WormLib is a open source python library for image analysis of *C. elegans* data. It incorporates a custom made Cellpose saegmentation model (ce-embryo), a Machine Learning classifier (WormForest) for cell ID and spot detection functions for spot detection in 3D (bigFISH). This modular pipeline for can used in batch automated image analysis.



### 2. Getting Started

#### 2.1 Create conda env and install packages

For users new to WormLib, follow this basic workflow:

1. Load Data: Specify directory, mRNA name, file type, and analysis functions via the Input tab
2. Segment Cells: Select segmentation type of interest in the Segmentation tab
3. Spot detection: Detect spots in the Tracking tab
4. Analyze Results: User-verification with plots and statistics in output directory
5. Export Data: Save results and visualizations via the Export tab

#### About WormLib

##### WormLib is a open source python library for image analysis of C. elegans data. It incorporates a custom made Cellpose saegmentation model (ce-embryo), a Machine Learning classifiers for cell ID and spot detection functions for spot detection in 3D (bigFISH). 

##### It provides a modular pipeline for automated image analysis that can used in batch.

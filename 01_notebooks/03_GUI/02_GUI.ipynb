```python
import os
import subprocess
import ipywidgets as widgets
from IPython.display import display

# Input directory selector
download_dir = widgets.Text(
    description="Download dir:",
    placeholder="/path/to/local_download_directory",
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='90%')
)

python_script_path = widgets.Text(
    description="Python script:",
    placeholder="/path/to/ce-bigfish.py",
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='90%')
)

# Image type dropdown
image_type = widgets.Dropdown(
    options=['DV_IMAGES', 'TIFF_IMAGES', 'N2D_IMAGES'],
    value='DV_IMAGES',
    description='Image type:',
    style={'description_width': 'initial'}
)

# Channel name inputs
cy5 = widgets.Text(value="set-3_mRNA", description="Cy5:")
mcherry = widgets.Text(value="erm-1_mRNA", description="mCherry:")
fitc = widgets.Text(value="PH-GFP", description="FITC:")
dapi = widgets.Text(value="DAPI", description="DAPI:")
brightfield = widgets.Text(value="brightfield", description="Brightfield:")

# Toggles for operations
toggle_options = {
    "RUN_EMBRYO_SEGMENTATION": widgets.Checkbox(value=True),
    "RUN_CELL_SEGMENTATION": widgets.Checkbox(value=True),
    "SPOT_DETECTION": widgets.Checkbox(value=True),
    "RUN_mRNA_HEATMAPS": widgets.Checkbox(value=True),
    "RUN_PROTEIN_HEATMAPS": widgets.Checkbox(value=True),
    "ANALYZE_RNA_DENSITY": widgets.Checkbox(value=True),
    "GENERATE_DONUT_MASK": widgets.Checkbox(value=True),
    "GENERATE_PGRANULE_MASK": widgets.Checkbox(value=True),
    "CALCULATE_MEMBRANE_COLOCALIZATION": widgets.Checkbox(value=True),
    "CALCULATE_NUCLEI_COLOCALIZATION": widgets.Checkbox(value=True),
    "CALCULATE_PGRANULE_COLOCALIZATION": widgets.Checkbox(value=True),
    "CALCULATE_mRNA_mRNA_COLOCALIZATION": widgets.Checkbox(value=True)
}

toggles_box = widgets.VBox([
    widgets.HBox([widgets.Label(k), toggle_options[k]]) for k in toggle_options
])

# Button to launch
launch_button = widgets.Button(description="Run ce-bigfish.py", button_style="success")

output = widgets.Output()

def run_script(_):
    with output:
        output.clear_output()
        print("Preparing to run script...")
        
        # Set environment variables
        os.environ["DV_IMAGES"] = "True" if image_type.value == "DV_IMAGES" else "False"
        os.environ["TIFF_IMAGES"] = "True" if image_type.value == "TIFF_IMAGES" else "False"
        os.environ["N2D_IMAGES"] = "True" if image_type.value == "N2D_IMAGES" else "False"

        os.environ["Cy5"] = cy5.value
        os.environ["mCherry"] = mcherry.value
        os.environ["FITC"] = fitc.value
        os.environ["DAPI"] = dapi.value
        os.environ["brightfield"] = brightfield.value

        for key, widget in toggle_options.items():
            os.environ[key] = "True" if widget.value else "False"

        # Prepare arguments
        folder = os.path.join(download_dir.value, "input")
        output_dir = os.path.join(download_dir.value, "output")
        
        # Example execution
        print("Running script...")
        try:
            subprocess.run(
                ["python", python_script_path.value, folder, output_dir],
                check=True
            )
        except subprocess.CalledProcessError as e:
            print("Script failed:", e)

launch_button.on_click(run_script)

# Display everything
display(widgets.VBox([
    download_dir,
    python_script_path,
    image_type,
    widgets.Label("Channel Names:"),
    cy5, mcherry, fitc, dapi, brightfield,
    widgets.Label("Select Processing Options:"),
    toggles_box,
    launch_button,
    output
]))
```
